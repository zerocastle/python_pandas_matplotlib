```python
import pandas as pd
import seaborn as sns # 아나콘다를 설치하면 자동적으로 설치 되는 라이브러리다 (시각화 툴이다.)
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
# 이런식으로 각 컬럼에 대해서 데이터를 불러올수 있다.
columns = ['sepal_length','sepal_width','petal_length']
df[columns].head() 
# 이런식으로 값을 뽑아 올수 있지만 컬럼 값 안에 한글이 있다면 안될 경우가 있다.
df.sepal_length

#그럴떄는 이런식으로 정해 주어야 한다.
df['sepal_length']

#regex 안에 정규 표현식이 들어가며 지금 나와 있는거는 컬럼 값에 '_'이라는 문자가 들어 있는 것에대해서 추출해준다.
#아래 몇몇 예시
df.filter(regex='_') #컬럼 값중에 _가 들어 있는 컬럼에 대한추출
df.filter(regex='length$')#length로 끝나는 컬럼 값에 대한 추출
df.filter(regex='^sepal') #sepeal로 시작하는 문자에대한 컬럼값을 추출

#:앞에5는 추출하는 행 값이고 ,뒤에는 추출할 프로덕션이다. (쫌 쓸만한 자식)
df.loc[:5,'sepal_width':'petal_width'] 
#0~4까지의 행을 출력을 하고 1~2까지의 프로덕션 값을 추출한다. ※':' 전체를 가지고 온다는 뜻이다.
df.iloc[:5,[1,3]] 


#복합 논리적인 요소 포함 예제
#이것또한 앞에가 행이 되고 , 뒤에는 열이 된다. 즉, sepal_length 시리얼에서 5보다 큰 값에 'sepal_length','sepal_width' 를 가지고 오라는 뜻이다.
df.loc[df['sepal_length'] > 5 , ['sepal_length','sepal_width']]



```
