```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2015', periods=1000)) #2015/01/01 부터 1000일 들어가는 값은 랜덤한값
s.head(5)
pd.DataFrame(s).head(5)
```

```python
s.cumsum().head(2)
s = s.cumsum()
```

```python
s.plot() #matplotlib 으로 출력을해줌
```

```python
r = s.rolling(window=60)
r.mean()
```

```python
#이동평균을 구할떄 쓰인다.
s.plot(style='k--')
r.mean().plot(style='k')
```

```python
df2 = pd.DataFrame(np.random.randn(1000,5),
                  index=pd.date_range('1/1/2015',periods=1000),
                   columns=['A','B','C','D','E']
                  )
#컬럼 5개를 추가한 것에 index는 2015/01/01 로부터 1000일 동안에 시리즈 값을 넣어준다.

df2.head(5)

df2.rolling(window=150).sum().plot(subplots=True) # 각각의 컬럼의 값들을 분리해준다. 60개씩 이동해서 계산을한다.
```

```python
df2.rolling(window=len(df2),min_periods=1).mean().plot()
```
