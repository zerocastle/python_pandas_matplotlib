```python
import pandas as pd
import seaborn as sns
import numpy as np #다차원 배열 값을 처리하는 라이브러리 이다.
```

```python
#==================================================== Summarize data =============================================
df = sns.load_dataset('iris')
df.shape #150 개의 행과 5개의 열을 가지고 있다고 알려줌
df['species'].value_counts() #petal_width가 가지고 있는 데이터들의 종류 별로 집계를 해줌

#이런식으로 많이 쓰인다. 각각의 행들에 값들의 집계를 뽑아 내어 통계를 낸것을 프레임화 시킨다.
pd.DataFrame(df['species'].value_counts())

#[0]은 행들을 말하고 [1]은 열들을 말한다.
df.shape[0]


#species라는 열에 유니클한 값들의 총계를 나타내준다.
df['species'].nunique()

df.describe(include=[np.object]) #가장많이 등장하는 프리퀀시를 보여준다.




```

```python
df['petal_width'].mean() # 편균값 이거 쓸꺼같다
df['petal_width'].max()
df['petal_width'].std() #표준편자구하기
df['petal_width'].quantile([0.25,0.76]) # 사이에 값을 보여주는건데 여기에 min 값과 max값을 집어넣어 출력하게 하면 될꺼 같다.
```
