```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {'A' : {0:'a',1:'b',2:'c'},
    'B':{0:1,1:3,2:5},
     'C':{0:2,1:4,2:6}
    }
)

df
```

```python
pd.melt(df,id_vars=['A'],value_vars=['B','C']) #열에 있는 값을 행으로 옮긴뒤 값이 대칭된다.
df2 = pd.melt(df,value_vars=['A','B','C']) # 값을 행으로 보낸다.
df2
```

```python
df2.pivot(columns='value',values='variable') #행에있는 데이터를 열로 보낸다.
```
