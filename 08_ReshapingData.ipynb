```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df.shape #mpg 데이터를 불어 들이고 행,열로 이루어져 있다는걸 알려준다
df.head()
df.sort_values('mpg').head() #mpg 기준으로 정렬을 해준다.

df.sort_values('mpg',ascending=True).head() #기본 값이 오름차순이기에 내림차순으로 정렬을 하려면 ascending=false로 해둔다.

```

```python
df.rename(columns = {'model_year' : 'year'}).head() #해당 컬럼에 이름을 변경한다.
```
